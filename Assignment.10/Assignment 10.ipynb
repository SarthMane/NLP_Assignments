```python
import nltk
from nltk.util import ngrams
from nltk.tokenize import word_tokenize
from collections import defaultdict, Counter
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
corpus = """
Natural language processing is a part of artificial intelligence.
Language models are used for translation and speech recognition.
Auto complete systems predict the next word in a sentence.
N gram models calculate probabilities of word sequences.
Artificial intelligence is transforming the world.
"""

tokens = word_tokenize(corpus.lower())

print(tokens[:20])
```

```python
# Create bigrams
bigrams = list(ngrams(tokens, 2))

# Count frequencies
bigram_freq = Counter(bigrams)

# Create conditional frequency table
model = defaultdict(Counter)

for w1, w2 in bigram_freq:
    model[w1][w2] += bigram_freq[(w1, w2)]

print("Sample Bigram Model:")
for word in list(model.keys())[:5]:
    print(word, "->", dict(model[word]))
```

```python
bigram_prob = defaultdict(dict)

for w1 in model:
    total_count = sum(model[w1].values())
    for w2 in model[w1]:
        bigram_prob[w1][w2] = model[w1][w2] / total_count

print("Sample Probabilities:")
for word in list(bigram_prob.keys())[:5]:
    print(word, "->", bigram_prob[word])
```

```python
def predict_next_word(input_text):
    input_tokens = word_tokenize(input_text.lower())

    if len(input_tokens) == 0:
        return "No input provided."

    last_word = input_tokens[-1]

    if last_word in bigram_prob:
        next_words = bigram_prob[last_word]
        predicted_word = max(next_words, key=next_words.get)
        return predicted_word
    else:
        return "No prediction available."
```

```python
test_sentences = [
    "artificial",
    "language",
    "auto",
    "speech"
]

for sentence in test_sentences:
    prediction = predict_next_word(sentence)
    print("Input:", sentence)
    print("Predicted Next Word:", prediction)
    print("-" * 40)
```
